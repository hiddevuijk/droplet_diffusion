```python
import numpy as np
import matplotlib.pyplot as plt
from sys import exit
```

```python
c = np.loadtxt("data/c.dat").T
drops = np.loadtxt("data/drops.dat")

Lx = 50
Ly = 25
Nx = 100
Ny = 50

dx = Lx / Nx
dy = Ly / Ny
aspect = dy / dx


fig = plt.figure()

ax1  = fig.add_axes([0.1, 0.075, 0.85, 0.3])
ax0  = fig.add_axes([0.1, 0.475, 0.85, 0.3])
axcb = fig.add_axes([0.1, 0.875, 0.85, 0.05])


ax = ax0
#im = ax.imshow(c, interpolation='none', origin='lower', vmin = 0.7, vmax = 0.8)
im = ax.imshow(c, interpolation='none', origin='lower')
ax.set_title("concentration")


fig.colorbar(im, cax=axcb, orientation='horizontal')


ax.set_aspect(aspect)

ax = ax1
ax.set_title("Droplets")

ax.set_xlim([0,Lx])
ax.set_ylim([0,Ly])
ax.set_aspect(1)

for i in range(drops.shape[0]):
    x = drops[i][0]
    y = drops[i][1]
    R = drops[i][2]

    if R > 0:
        #print(x,y,R)
        circle = plt.Circle((x, y), R) 
        circle.set_alpha(0.3) 
        ax.add_artist(circle)


 
plt.show()
```

```python
Rlist = []
for i in range(drops.shape[0]):
    R = drops[i][2]
    if R > 0:
        Rlist.append(R)
    
h, bins = np.histogram(Rlist, density=True, bins=10)
db = bins[1] - bins[0]
bins += db /2.0
x = bins[:-1]
plt.scatter(x,h)
plt.xlim([0,1.1*bins[-1]])
plt.show()
```
